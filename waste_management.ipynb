```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/waste_sensor_data.csv')
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.dropna()
```

```python
data.drop_duplicates()
```

```python
waste_counts = data['waste_type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(waste_counts, labels=waste_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Waste Types')
plt.axis('equal')

plt.show()
```
